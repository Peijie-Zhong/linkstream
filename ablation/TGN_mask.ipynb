```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


src_feat = np.load("/Users/acw721/Desktop/research/linkstream/results/node_event_emb/full_src.npy")
dst_feat = np.load("/Users/acw721/Desktop/research/linkstream/results/node_event_emb/full_dst.npy")
meta = pd.read_csv("/Users/acw721/Desktop/research/linkstream/results/node_event_emb/full_meta.csv")

X = np.concatenate([src_feat, dst_feat], axis=0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

K = 5
kmeans = KMeans(n_clusters=K, n_init=20, random_state=0)
labels = kmeans.fit_predict(X_scaled)

N = len(meta)
src_labels = labels[:N]
dst_labels = labels[N:]


out_df = pd.DataFrame({
    "source": meta["source"],
    "destination": meta["destination"],
    "timestamp": meta["timestamp"],
    "source_commu": src_labels,
    "destination_commu": dst_labels,
})

out_df.to_csv("/Users/acw721/Desktop/research/linkstream/results/p0.8_mu0.2_1/TGN_mask_cluster.csv", index=False)
print(out_df.head())
```
