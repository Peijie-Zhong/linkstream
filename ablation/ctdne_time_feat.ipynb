```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans

src_feat = np.load("/Users/acw721/Desktop/research/linkstream/pretrain/node_event_feat_64/src_feat_64.npy")
dst_feat = np.load("/Users/acw721/Desktop/research/linkstream/pretrain/node_event_feat_64/dst_feat_64.npy")
meta = pd.read_csv("/Users/acw721/Desktop/research/linkstream/pretrain/node_event_feat_64/metadata.csv")

X = np.concatenate([src_feat, dst_feat], axis=0)   # [2N, D]
print(X.shape)
```

```python
K = 5

kmeans = KMeans(n_clusters=K, n_init=20, random_state=0)
labels = kmeans.fit_predict(X)

print(labels.shape)   # [2N]
```

```python
N = len(meta)

src_labels = labels[:N]
dst_labels = labels[N:]

meta["src_cluster"] = src_labels
meta["dst_cluster"] = dst_labels

meta.to_csv("node_event_cluster_64.csv", index=False)
print(meta.head())
```

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


src_feat = np.load("/Users/acw721/Desktop/research/linkstream/pretrain/node_event_feat_64/src_feat_64.npy")
dst_feat = np.load("/Users/acw721/Desktop/research/linkstream/pretrain/node_event_feat_64/dst_feat_64.npy")
meta = pd.read_csv("/Users/acw721/Desktop/research/linkstream/pretrain/node_event_feat/metadata.csv")

X = np.concatenate([src_feat, dst_feat], axis=0)

# 标准化后做 KMeans
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

K = 5
kmeans = KMeans(n_clusters=K, n_init=20, random_state=0)
labels = kmeans.fit_predict(X_scaled)

# 拆回 source / destination
N = len(meta)
src_labels = labels[:N]
dst_labels = labels[N:]


out_df = pd.DataFrame({
    "source": meta["source"],
    "destination": meta["destination"],
    "timestamp": meta["timestamp"],
    "source_commu": src_labels,
    "destination_commu": dst_labels,
})

out_df.to_csv("node_event_cluster_64.csv", index=False)
print(out_df.head())
```
